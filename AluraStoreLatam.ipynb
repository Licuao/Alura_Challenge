### Importación de datos



```python
simport pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
tiendas = [tienda, tienda2, tienda3, tienda4]

for i, df in enumerate(tiendas):
    ingreso_total = df['Precio'].sum()
    print(f"Ingreso total de Tienda {i+1}: {ingreso_total}")
```

# 2. Ventas por categoría

```python
for i, df in enumerate(tiendas):
    ventas_por_categoria = df['Categoría del Producto'].value_counts()
    print(f"\nVentas por categoría en Tienda {i+1}:")
    print(ventas_por_categoria)
    print(f"Categoría más popular en Tienda {i+1}: {ventas_por_categoria.index[0]} con {ventas_por_categoria.iloc[0]} ventas")
```

# 3. Calificación promedio de la tienda


```python
tiendas = [tienda, tienda2, tienda3, tienda4]

for i, df in enumerate(tiendas):
    calificacion_promedio= df['Calificación'].mean()
    print(f"clasificacion Promedio de la tienda: {i+1}: {calificacion_promedio}")
```

# 4. Productos más y menos vendidos

```python
for i, df in enumerate(tiendas):
    ventas_por_producto = df['Producto'].value_counts()

    most_sold_product = ventas_por_producto.index[0]
    most_sold_count = ventas_por_producto.iloc[0]

    least_sold_product = ventas_por_producto.index[-1]
    least_sold_count = ventas_por_producto.iloc[-1]

    print(f"\nTienda {i+1}:")
    print(f"  Producto más vendido: {most_sold_product} ({most_sold_count} ventas)")
    print(f"  Producto menos vendido: {least_sold_product} ({least_sold_count} ventas)")
```

# 5. Envío promedio por tienda

```python
for i, df in enumerate(tiendas):
    costo_envio_promedio = df['Costo de envío'].mean()
    print(f"Costo de envío promedio de Tienda {i+1}: {costo_envio_promedio}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

all_category_sales = {}
for i, df in enumerate(tiendas):
    store_name = f'Tienda {i+1}'
    category_sales = df['Categoría del Producto'].value_counts()
    all_category_sales[store_name] = category_sales

category_sales_df = pd.DataFrame(all_category_sales).fillna(0)
store_names = [f'Tienda {i+1}' for i in range(len(tiendas))]
category_sales_df = category_sales_df[store_names]
plt.figure(figsize=(15, 8))
category_sales_df.T.plot(kind='bar', stacked=True, figsize=(15, 8))
plt.title('Ventas por Categoría de Producto por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Número de Ventas')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Categoría del Producto', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
